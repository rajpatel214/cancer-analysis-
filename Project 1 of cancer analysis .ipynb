```python
x=data.drop("diagnosis",axis=1)

y=data["diagnosis"]
```

# train_test_split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

# scaling

```python

from sklearn.preprocessing import StandardScaler

r=StandardScaler()

train_sacling=r.fit_transform(x_train)

test_sacling=r.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score,KFold

list=[LogisticRegression,DecisionTreeClassifier,RandomForestClassifier,SVC,KNeighborsClassifier]

k=KFold(n_splits=5)

#without scaling data 
for i in list:
    model=i()
    print(i)
    print(cross_val_score(model,x_train,y_train,cv=k).mean())
```

```python
#with scaling data 

for i in list:
    model=i()
    print(i)
    print(cross_val_score(model,train_sacling,y_train,cv=k).mean())

```

# LogisticRegression

```python
#LogisticRegression  high score give from all those model
lr=LogisticRegression(C=50,random_state=10)

lr.fit(train_sacling,y_train)

lr.score(train_sacling,y_train)

lr.score(test_sacling,y_test)
```

# RandomForestClassifier

```python
r=RandomForestClassifier(bootstrap=False, max_depth=5, min_samples_split=100,
                       n_estimators=200, random_state=30)

r.fit(train_sacling,y_train)

r.score(train_sacling,y_train)

r.score(test_sacling,y_test)
```
